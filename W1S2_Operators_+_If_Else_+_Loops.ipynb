## **1. Arithmetic Operators**

Used for math operations.

```python
# Addition
print(5 + 6)
```

```python
# Subtraction
print(5 - 6)
```

```python
# Multiplication
print(5 * 6)
```

## **2. Relational / Comparison Operators**

Used to compare values.

```python
# Greater than
print(4 > 5)
```

```python
# Less than
print(4 < 5)
```

```python
# Greater than or equal to
print(4 >= 4)
```

```python
# Less than or equal to
print(4 <= 4)
```

```python
# Equal to → True
print(4 == 4)
```

```python
# Not equal to
print(4 != 4)
```

## **3. Logical Operators**

Used to combine conditions.

```python
# and returns the first false value
print(1 and 0)
```

```python
# or returns the first true value
print(1 or 0)
```

```python
# not reverses the truth value
print(not 1)
```

## **4. Bitwise Operators**

Work on bits (0/1).

```python
# Bitwise AND (compares bits of both numbers)
print(2 & 3)
```

```python
# Bitwise OR (sets bit if any bit is 1)
print(2 | 3)
```

```python
# Bitwise XOR (sets bit if bits are different)
print(2 ^ 3)
```

```python
# Bitwise NOT (inverts all bits)
print(~3)
```

```python
# Right shift (moves bits right, divides by 2)
print(4 >> 2)
```

```python
# Left shift (moves bits left, multiplies by 2)
print(5 << 2)
```

## **5. Assignment Operators**

Used to assign values.

```python
# Assignment operator (=) assigns value
a = 2

# += adds value to existing variable (a = a + 2)
a += 2

# Print final value of a
print(a)
```

## **6. Membership Operators**

Check presence in a sequence.
*  in - checks if a value exists in a sequence
*  not in - checks if a value does NOT exist in a sequence

```python
# 'in' checks if a value exists in a sequence
print('D' in 'Delhi')
```

```python
# 'not in' checks if a value does NOT exist in a sequence
print('D' not in 'Delhi')
```

```python
# Check if 1 is present in the list
print(1 in [2, 3, 4, 5, 6])
```

```python
# Check if 1 is NOT present in the list
print(1 not in [2, 3, 4, 5, 6])
```

# **If-else in Python**

if-else is a conditional control statement used to execute code based on a condition.

The condition is an expression that evaluates to True or False.

If the condition is True, the if block runs; otherwise, the else block runs.

Python uses indentation (not braces) to define blocks.

## **Branching**

Branching is when a program chooses between multiple paths of execution based on conditions.

Think of it like a road with forks -> depending on the condition, the program takes a different path.

*  Branching allows decision-making in multiple directions.

*  Only one path executes per if–elif–else block.

*  Nested branching helps handle complex decisions.

*  Proper indentation is critical.


## **1. if–else**

if–else is used to make decisions in Python.
The program checks a condition:

*  If the condition is True → run the if block

*  If the condition is False → run the else block

## **2. if–elif–else**

When there are multiple conditions

Use it when you need your program to choose between more than two possibilities.

*  if → checks the first condition

*  elif → checks other conditions if previous if/elif is False

*  else → runs when all previous conditions are False

## **3. Nested if–else**

A nested if–else is an if–else statement inside another if or else.

It is used when a decision depends on another decision.

*  Nested if–else allows complex decisions.

*  Can have multiple levels of nesting, but avoid too many levels (hard to read).

*  Indentation is very important; wrong indentation gives syntax errors.

### **While Loop**

Executes a block of code as long as a condition is True.
Used when number of iterations is not known in advance.

*  Condition is checked before each iteration

*  Loop stops when condition becomes False

*  Avoid infinite loops by changing variables inside the loop

```python
# While loop example-> program to print the table

number = int(input('Enter the number: '))
i = 1
while i<11:
  print(number,'*',i,'=',number * i)
  i += 1
```

```python
 # while loop with else

x = 1
while x < 3:
  print(x)
  x += 1
else:
  print('limit crossed')
```

#### Guessing game

```python
# generate a random integer between 1 and 100
import random
jackpot = random.randint(1,100)

guess = int(input('guess karo'))
counter = 1
while guess != jackpot:
  if guess < jackpot:
    print('galat!guess higher')
  else:
    print('galat!guess lower')

  guess = int(input('guess karo'))
  counter += 1

else:
  print('correct guess')
  print('attempts',counter)
```

*  range(start, stop, step) is commonly used for numbers

```python
# For loop demo

for i in {1,2,3,4,5}:
  print(i)
```

> Current Year Population = x * 1.1

In this equation, the current year's population is equal to the previous year's population multiplied by 1.1, representing a 10% increase.

To find the population of the previous year (x), we can rearrange the equation as follows:

> x = Current Year Population / 1.1

Using this simplified equation, if you have the current year's population (e.g., 10,000), you can divide it by 1.1 to calculate the population of the previous year.

This equation allows you to calculate the population for each year, assuming you know the population of the current year and want to find the population of the previous year.

## **Nested Loops**

A nested loop means a loop inside another loop.

The inner loop runs completely for each iteration of the outer loop.

**How it Works**

*  Outer loop runs first

*  For each outer loop iteration, the inner loop runs fully

**Nested Loop – Real-Life Example (Classroom)**


*  Outer Loop: Har classroom me jaana

*  Inner Loop: Us classroom ke sab bachon ke naam lena

Simple Understanding

*  Pehle classroom select hota hai (outer loop)

*  Phir us classroom ke har bache ka naam liya jata hai (inner loop)


**When to Use Nested Loops**

*  Pattern printing

*  Working with 2D lists / matrices

* Comparing elements (pair-wise)

*  Menu & sub-menu systems

**Important Points**

*  Inner loop executes more times

*  More nesting = more time complexity

*  Avoid deep nesting if possible

```python
# Example --> Unique Pairs

for i in range(1,5):
  for j in range(1,5):
    print(i,j)
```

## **Loop Control Statement**

Loop control statements allow you to control the execution flow of loops.

Python provides three main loop control statements:

*  break → exits the loop immediately

*  continue → skips the current iteration

*  pass → placeholder, no operation

**Important Notes**

*  break and continue only affect the current loop

*  pass is not loop-specific; it works anywhere

*  Avoid excessive use of break to keep code readable

### **1. break Statement**

The break statement is used to terminate the loop immediately, even if the loop condition is still true.

**How it Works**

*  When break is executed, the loop stops completely.

*  Control moves to the statement after the loop.

**Real Life Example using break - Linear Search**

In linear search, break is used to stop the loop immediately after the element is found, just like stopping the search once the correct student is identified.

```python
for i in range(1,10):
  if i == 5:
    break
  print(i)
```

#### Write a Python program to print all prime numbers between a given lower range and upper range (both inclusive) entered by the user, using nested loops and the break statement.

```python
lower = int(input('Enter lower range: '))
upper = int(input('Enter upper range: '))

for i in range(lower,upper+1):
  for j in range(2,i):
    if i % j == 0:
      break
  else:
    print(i)
```

### **2. continue Statement**

The continue statement is used to skip the current iteration and move to the next iteration of the loop.

**How it Works**

*  The remaining code in the current loop iteration is skipped.

*  Loop does not stop, only the current cycle is skipped.

**Real-Life Example of continue (E-Commerce – Skip Product)**

You are browsing products on an e-commerce website (like Amazon/Flipkart).

*  The system checks products one by one

*  If a product is out of stock, it is skipped

*  The website continues showing the next available product

*  This skipping behavior is done using continue.

**Use Cases:**

*  Skipping unwanted values

*  Ignoring invalid inputs

*  Filtering data inside loops

```python
for i in range(1,10):
  if i == 5:
    continue
  print(i)
```

### **3. pass Statement**

The pass statement is a null statement. It does nothing and is used when a statement is required syntactically but no action is needed.

**How it Works**

*  Python expects a statement in blocks like if, for, while, def, class.

*  pass prevents syntax errors.

**Use Cases:**

*  Creating empty blocks

*  Writing code structure first

*  Placeholder for future logic

```python
for i in range(1,100):
  pass
```
